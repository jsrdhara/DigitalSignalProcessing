# Question-1

```python
import numpy as np
import matplotlib.pyplot as plt

def fxNthOrderPolyDTSignal(roots,tmin,tmax,noOfSamples):
    
    N = noOfSamples-1
    T = (tmax-tmin)/N
    coeff = np.poly(roots)
    X_val=np.arange(tmin,tmax+T,T)
    y_val =[]
    n = []
    for t in X_val:
        y_val.append(np.polyval(coeff,t))
        n.append((t-tmin)/T)
    plt.subplot(3,1 , 1)
    fig = plt.gcf()
    fig.set_size_inches(10, 10)
    plt.ylabel('x(t)')
    plt.xlabel('Time t')
    plt.plot(X_val,y_val,'-')
    plt.subplot(3, 1, 2)
    plt.xlabel('Sample index n')
    z = np.zeros(noOfSamples)
    plt.plot(n, z, 'o')
    plt.subplot(3, 1, 3)
    plt.ylabel('x(n)')
    plt.xlabel('sample index n')
    plt.stem(n,y_val)
    plt.show()
    print(' vector of time points at which samples were drawn')
    print([round(i) for i in X_val])
    print('\n')
    print('vector of sample indices i.e. n')
    print([round(i) for i in n])
    print('\n')
    print('vector of sampled signal i.e. x[n]')
    print([round(i) for i in y_val])
    
    
fxNthOrderPolyDTSignal(roots = [8,1,-5],tmin=-5,tmax=5,noOfSamples=41)
```

# Question 2

```python
def step_operation(k):
    x_u=[]
    y_u=[]
    for i in range(-30,30):
        x_u.append(i)
        if i>=0:
            y_u.append(1)
        else:
            y_u.append(0)
    
    x_k=[]
    y_k=[]
    for j in range(-30,30):
        x_k.append(j)
        if j>=k:
            y_k.append(1)
        else:
            y_k.append(0)
            
    g_u = np.array(y_u)-np.array(y_k)
    fig = plt.gcf()
    fig.set_size_inches(10, 10)
    plt.subplot(3, 1, 1)
    plt.title('u(n)')
    plt.stem(x_u,y_u)
    plt.subplot(3, 1, 2)
    plt.title('u(n-k)')
    plt.stem(x_k,y_k)
    plt.subplot(3, 1, 3)
    plt.title('g(n)=u(n)-u(n-k)')
    plt.stem(x_k,g_u)
    plt.show()
    
step_operation(k=4)
```

# T_duration = n_duration*dt

# T_pause = n_pause*dt

# In the given question :

# Fs = 8000 = 1/Ts
 
# Ts = 0.000125

# n_duration = T_duration/dt = 1600

# n_pause = T_pause/dt = 160


```python
def ss_dtmf (number, dt,n_duration, n_pause):
    Fs = 8000
    Ts = 1/Fs
    n_duration_array = np.arange(0,n_duration)
    n_pause_array = list(np.zeros(n_pause))
    Total_list = []
    for n in number:
        if n==1:
            f1 = 697
            f2 = 1209
        elif n==2:
            f1 = 697
            f2 = 1336
        elif n==3:
            f1 = 697
            f2 = 1477
        elif n==4:
            f1 = 770
            f2 = 1209
        elif n==5:
            f1 = 770
            f2 = 1336
        elif n==6:
            f1 = 770
            f2 = 1477
        elif n==7:
            f1 = 852
            f2 = 1209
        elif n==8:
            f1 = 852
            f2 = 1336
        elif n==9:
            f1 = 852
            f2 = 1477
        elif n==10:
            f1 = 941
            f2 = 1209
        elif n==0:
            f1 = 941
            f2 = 1336
        elif n==11:
            f1 = 941
            f2 = 1477
        elif n==12:
            f1 = 0
            f2 = 0
            
        X_t = []
        
        for i in n_duration_array:
            X_t.append(np.sin(2*np.pi*f1*i*dt) + np.sin(2*np.pi*f2*i*dt))
            
        X_t = X_t+n_pause_array
        
        Total_list=Total_list+X_t
    n_samples = np.arange(0,len(Total_list))
    fig = plt.gcf()
    fig.set_size_inches(20, 8)
    plt.plot(n_samples,Total_list)
    plt.show()
    return Audio(np.array(Total_list), rate=8000)
```
